```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
f1 = lambda x: x**2 + 1
tt = np.linspace(-4,6,200)
```

```python
plt.plot(tt,f1(tt))
plt.title('Original function')
plt.axvline(x = 2, color = 'red', linestyle = 'dashed')
plt.axvline(x = 4, color = 'red', linestyle = 'dashed')
```

```python
mult = 5 #lambda
g1 = lambda x: (x-2)*(x-4)
ff = f1(tt) + mult * g1(tt)
```

```python
plt.plot(tt,ff)
plt.title('Lambda = 100')
plt.axvline(x = 2, color = 'red', linestyle = 'dashed')
plt.axvline(x = 4, color = 'red', linestyle = 'dashed')
```

```python
plt.plot(tt,ff)
plt.title('Lambda = 20')
plt.axvline(x = 2, color = 'red', linestyle = 'dashed')
plt.axvline(x = 4, color = 'red', linestyle = 'dashed')
```

```python
plt.plot(tt,ff)
plt.title('Lambda = 5')
plt.axvline(x = 2, color = 'red', linestyle = 'dashed')
plt.axvline(x = 4, color = 'red', linestyle = 'dashed')
```

```python
//define dual function
mult = np.linspace(0,100,200)
ind = 0
dual = np.zeros(mult.size)
for m in mult:
    dual[ind] = np.min(f1(tt)+m*g1(tt))
    ind = ind + 1
```

```python
plt.plot(mult,dual)
plt.title('Dual Function')
```

```python
print(np.max(dual), np.argmax(dual))
```

$\lambda^* = 4$ and $d(\lambda^*) = 5$

Strong duality holds, as the max of our dual function(5), is equal to the min of our constrained problem

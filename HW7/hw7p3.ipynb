```python
import numpy as np
import cvxpy as cp

# Construct the problem
N = 1000 # Population size
S = 20 # Number of infected individuals
M = 100 # Number of tests
K = 5 # Number of splits of each sample

# Define x0
ind0 = np.random.choice(N,S,0) # index subset 
x0 = np.zeros(N) 
x0[ind0] = np.random.rand(S)

# Define A
A = np.zeros((M,N))
for i in np.arange(N):
    ind = np.random.choice(M,K,0)
    A[ind,i] = 1

y = A @ x0
```

```python
x = cp.Variable(1000)
print(x)
prob = cp.Problem(cp.Minimize(cp.atoms.norm1(x)),[x >= 0, y == A @ x])
prob.solve()
print("Optimal value", prob.solve())
print("Optimal var")
print(x.value) # A numpy ndarray.
```

```python
#set a threshold value and set everything below to 0;
#this allows close to zero values to be ignored
sol = [0 if  i <= 1e-2 else 1 for i in x.value]
```

```python
#compare identified individuals
for i in range(1000):
    if sol[i] == 1: print("sol: ",i)
    if x0[i] > 0:print("x0: ",i)
    
```

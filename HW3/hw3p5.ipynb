```python
from matplotlib import pyplot
import numpy as np

def graddescent1(H, b): #version 1 gradient descent based on psuedocode in notes
    x = []
    num_rows, num_cols = b.shape
    x.append(np.zeros((num_rows, 1)))
    k = 0
    r = []
    r.append(b - np.matmul(H,x[k]))
    #print(x[0], r[0])
    while np.linalg.norm(r[k], 2) >= epsilon and k <= maxit:
        a = np.matmul(np.transpose(r[k]), r[k]) / np.matmul(np.matmul(np.transpose(r[k]), H), r[k])
        a = a.item()
        x.append(x[k] + a * r[k])
        r.append(r[k] - a * np.matmul(H,r[k]))
        k = k + 1
    print("Number of iterations: ", k)
    return x

def graddescentfs (H, b, a): #fixed step gradient descent
    x = []
    num_rows, num_cols = b.shape
    x.append(np.zeros((num_rows, 1)))
    k = 0
    r = []
    r.append(b - np.matmul(H,x[k]))
    while np.linalg.norm(r[k], 2) >= epsilon and k <= maxit:
        x.append(x[k] + a * r[k])
        r.append(r[k] - a * np.matmul(H,r[k]))# r.append(b - np.matmul(H,x[k+1]))
        k = k + 1
    print("Number of iterations: ", k)
    return x
```

```python
#declaring global variables for precision and max iterations
global epsilon 
epsilon = 10 ** -4 #precision of 4 decimal places
global maxit
maxit = 200
```

```python
#part a
A = np.matrix([[1, 0],[0, 1]])
y = np.matrix([[1],[1]])
b = np.matmul(np.transpose(A),y)
H = np.matmul(np.transpose(A), A)
a = 1 #fixed step size

print(graddescent1(H , b))
print(graddescentfs(H, b, a))
#note that best step size is a=1 for this system
```

```python
#part b
A = np.matrix([[1, .99],[.99, 1]])
y = np.matrix([[1.99],[1.99]])
b = np.matmul(np.transpose(A),y)
H = np.matmul(np.transpose(A), A)
a = .1 #fixed step size

eig = np.linalg.eigvals(H)
k = max(eig)/min(eig)
print("k = ", k)
print(graddescent1(H , b))
print(graddescentfs(H, b, a))

#This is unexpected, as version 1 requires only 1 iteration, even though k is very large and fixed step
#only requires 40 iterations at a=.1
```

```python
#part c
epsilon = 10 ** -9 #set precision to 9 decimal points to make gradient descent v1 go to the proper solution
A = np.matrix([[1, .99],[1, 1.01]])
y = np.matrix([[1.99],[2.01]])
b = np.matmul(np.transpose(A),y)
H = np.matmul(np.transpose(A), A)
a = .2 #fixed step size

eig = np.linalg.eigvals(H)
k = max(eig)/min(eig)
print("k = ", k)
print(graddescent1(H , b))
print(graddescentfs(H, b, a))
#the matrix is no longer symmetric, and k increases slightly
```

To get a more exact solution, $\epsilon$ must be smaller so that the algorithm doesn't stop after its first iteration at an inoptimal solution

```python
#part e
epsilon = 10 ** -9
A = np.matrix([[2, -1],[0, -1]])
y = np.matrix([[1],[-1]])
b = np.matmul(np.transpose(A),y)
H = np.matmul(np.transpose(A), A)
a = .2 #fixed step size

eig = np.linalg.eigvals(H)
k = max(eig)/min(eig)
print("k = ", k)
print(graddescent1(H , b))
print(graddescentfs(H, b, a))
```

At a precision of 9 decimal places, v1 takes 61 iterations to reach a solution and fixed step takes 126 iterations at step size of .2. With the max iterations set to 200, in part c the algorithm never reaches the desired precision within the iteration bound. As $\kappa$ is still > 2, the change between iterations is still inoptimal, but the initial estimation for x, $x_0$ is computed as $\begin{bmatrix}2\\0\end{bmatrix}$ for d and $\begin{bmatrix}4\\4\end{bmatrix}$ for c, which is collinear with the actual solution of $\begin{bmatrix}1\\1\end{bmatrix}$, so it would make sense that linear regression that optimizes step size would reach a solution faster for c than d.
